```python
import numpy as np
import pandas as pd
```

# read the dataset

```python
dataset=pd.read_csv("bikebuyer.csv")
```

```python
dataset
```

```python
dataset.head(5)
```

```python
dataset.isnull().any()
```

```python
dataset.isnull().sum()
```

```python
dataset["ID"].fillna(dataset["ID"].mean(),inplace=True)
dataset["Marital Status"].fillna(method="ffill",inplace=True)
dataset["Gender"].fillna(method="ffill",inplace=True)
dataset["Children"].fillna(dataset["Children"].mean(),inplace=True)
dataset["Commute Distance"].fillna(dataset["Commute Distance"].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
x1 = dataset.drop(['Bike Buyer'],axis=1)
```

```python
y1 = dataset[['Bike Buyer']]
```

```python
y1 
```

```python
k=list(dict(x1.dtypes[x1.dtypes == 'object']).keys())
```

```python
k
```

```python
x1[k]
```

```python
z=pd.get_dummies(x1[k],drop_first=True)
```

```python
x = pd.concat([x1,z],axis=1)
```

```python
x.head()
```

```python
x.drop(k,axis=1,inplace=True)
```

```python
x
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y1,test_size = 0.2,random_state = 0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr =LogisticRegression()
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```
